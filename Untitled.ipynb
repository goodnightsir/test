```python
import matplotlib.pyplot as plt
import seaborn as sns

import matplotlib
zhfont= matplotlib.font_manager.FontProperties(fname=r'C:\Users\14744\Desktop\论文\毕业论文\最终数据\SimHei\SimHei.ttf',size=14)
sns.set(font='simhei')


group_labels = {
    1: '控制组',
    2: '道德规范组',
    3: '真实案例组',
    4: '数据算法组'
}
df['group_label'] = df['group'].map(group_labels)

# 设置画图的风格为白色背景

sns.set_style("whitegrid")

# 创建箱线图，调整为黑白灰色调
plt.figure(figsize=(10, 6))
boxplot = sns.boxplot(x='group_label', y='AA', data=df,palette = 'grey')

# 设置图表的标题和坐标轴标签
plt.title('Group wise AA Value Boxplot')
plt.xlabel('中文',fontproperties=zhfont)
plt.ylabel('AA Values')

# 显示图表
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib
plt.rcParams['font.sans-serif'] = ['KaiTi']  # 设置中文支持字体
plt.rcParams['axes.unicode_minus'] = False  # 正确显示负号
# 假设数据已经加载到DataFrame中，这里创建一个示例DataFrame
data = {
    'group': ['控制组','道德规范组', '真实案例组', '数据算法组','控制组','道德规范组', '真实案例组', '数据算法组','控制组','道德规范组', '真实案例组', '数据算法组'],
    'CF': [-1.33, -1.33,-1.33,-1.33,0,0,0,0,1.33,1.33,1.33,1.33],
    'M': [4.0463, 4.2472, 4.4209,4.4986,3.9499, 4.4042,4.4696,4.6271,3.8535,4.5613,4.5184,4.7556]
}
df = pd.DataFrame(data)

# 为了区分线条，分配不同的颜色和样式
colors = ['black', 'darkgrey', 'lightgrey', 'dimgray']
linestyles = ['-', '--', '-.', ':']
markers = ['o', 's', '^', 'x']

plt.figure(figsize=(10, 6))
plt.gca().set_facecolor('white')  # 设置背景颜色为白色

# 对于DataFrame中的每个组别，绘制一条线
for (i, group) in enumerate(df['group'].unique()):
    subset = df[df['group'] == group]
    plt.plot(subset['CF'], subset['M'], label=group, color=colors[i], linestyle=linestyles[i], marker=markers[i])

# 设置图例
plt.legend(fontsize = 12)

# 设置标题和轴标签
# plt.title('Group-wise CF vs. M')
plt.xlabel('算法熟悉度',fontsize = 16)
plt.ylabel('信息公平感',fontsize = 16)

# 显示图形
plt.show()
```
